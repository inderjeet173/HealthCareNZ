#install the libraries and import them 


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

# Load CVS Data 

```python
df=pd.read_csv('NZ_Health_Dataset.csv')
df.head()
```

# Quick Analysis

```python
# case by disease
df.groupby('Disease')['Cases'].sum()
# case by Region
df.groupby('Region')['Cases'].sum()
# case by Year
df.groupby('Year')['Cases'].sum()
# case by Age Group
bins = [0, 18, 35, 50, 100]
labels = ['0-18', '19-35', '36-50', '51+']
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels)
df.groupby('Age_Group')['Cases'].sum()
# case by Gender
df.groupby('Gender')['Cases'].sum()
```

# Matplotlib / Seaborn Visuals Inline

```python
sns.set(style="whitegrid")
plt.figure(figsize=(8, 5))
sns.countplot(x='Disease', data=df)
plt.title('Cases by Disease')
plt.show()
```

```python
pivot = df.pivot_table(index='Region', columns='Disease', values='Cases', aggfunc='sum', fill_value=0)
plt.figure(figsize=(10,6))
sns.heatmap(pivot, annot=True, fmt="d", cmap="YlOrRd")
plt.title('Cases by Region and Disease')
plt.show()
```
